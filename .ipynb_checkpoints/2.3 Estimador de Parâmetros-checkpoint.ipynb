# Estimador de Parâmetros

Inspirado no Notebook: https://github.com/jckantor/ND-Pyomo-Cookbook/blob/master/notebooks/07.01-Parameter-Estimation-Catalytic-Reactor.ipynb
Keywords: parameter estimation, ipopt usage, data fitting, tfest, procest, control, python <br />
Palavras chave: estimação de parâmetros, ipopt, ajuste linear e não linear, tfest, procest, matlab, controle

Este notebook demonstra a estimação de parâmetros de um motor CC acionando uma carga mecânica constante e um controlador PD, conforme descrito em: https://github.com/dbr-ufs/controle/blob/main/system_identification-final.pdf 



onde: $\omega$ = Velocidade do motor, 
$\it{V}$ = Tensão aplicada na armadura,
$\it{Ia}$ = corrente de armadura,
$\it{L}$ = Indutância da armadura,
$\it{e_a}$ = Forca contra eletromotriz,
$\it{R}$ = Resistência do enrolamento da armadura,
$\it{T}$ = Torque interno do motor = $k ia$,
$\it{J}$ = Momento de inercia da carga,
$\it{f}$ = Atrito viscoso no eixo.

Tem-se desta forma as duas equações que descrevem o sistema para a parte elétrica
e mecânica, sendo elas:

\begin{equation}
V(t)-R\cdot Ia(t) - L\frac{\mathrm{d} Ia(t)}{\mathrm{d} t}-e_a(t) = 0
\label{eq:motor_elet}
\end{equation}

\begin{equation}
J \dot{\omega}(t)= T-f\omega(t)
\label{eq:motor_mec}
\end{equation}

A conversão eletromecânica ocorre na armadura, onde se iguala a potência interna desenvolvida pelo motor:

\begin{equation}
P_{int} = e_a Ia = T \omega(t)
\end{equation}

A partir destas equações diferenciais pode-se chegar ao sistema de equações de estado 
descrito em:

\begin{equation}
\begin{matrix}
\begin{bmatrix}
\dot{Ia}\\ 
\dot{\omega} \\
\dot{\theta}
\end{bmatrix}= &
\begin{bmatrix}
-R/L & -k/L & 0\\ 
k/J & -f/J & 0 \\
0 & 1 & 0
\end{bmatrix}&
\begin{bmatrix}
Ia\\ 
\omega \\
\theta
\end{bmatrix} + &
\begin{bmatrix}
1\\ 
0\\
0
\end{bmatrix} V
\end{matrix}
\label{eq:motor_eq_estados}
\end{equation}

Com esse sistema de equações de estado, é possível obter-se as funções de transferências,
que descrevem o sistema, para uma entrada e uma saída. Encontra-se, portanto a função de transferência do motor, que relaciona a posição angular com a tensão de entrada do motor:

\begin{equation}
G(s)=\frac{k/J_L}{s(s+f/L)(s+R/L)+s\;k^2/J_L}
\label{eq:motor_pos_tf}
\end{equation}

Devido a dinâmica do sistema mecânico ser muito mais lenta que a dinâmica do sistema elétrico, 
pode-se desconsiderar a influencia do polo elétrico do sistema, ficando a função de transferência
assim:

\begin{equation}
G(s)=\frac{\frac{k}{J_L}}{s(s+\frac{f_L+k_1 k_2}{J_L})}
\nonumber
\end{equation}

A equação acima representa o comportamento da posição angular do motor baseado na entrada de tensão aplicada sobre seus terminais.

Para uma entrada em rampa, a coleta de dados foi realizada utilizando o seguinte controlador PD:

$PD(s) = k_p + k_d.s$.

Fechando a malha da realimentação unitária, observe que a função de transferência de malha fechada tem o seguinte formato: 

$T(s) = \frac{T_z.s+1}{T_w^2.s^2+2.\zeta.T_w.s+1}$, 

onde $T_z = \frac{k_d}{k_p}$ (devido ao controlador PD). Cabe a nós, portanto, identificar os demais parâmetros $\zeta$ e $T_w$.

## Passo 3. Definir a função objetivo

Para isso, simplesmente foi escolhido a diferência entre o valor estimado e o valor medido. Aplicando a função control.forced_response(sys,t,u) que faz a convolução da função de transferência com a entrada, encontra-se $y_e$ (valor estimado de y).

```python
kp = 3      # ganho proporcional
kd = 0.05   # ganho derivativo
Tz = kd/kp

def residuals(parameters, df):
    Zeta, Tw = parameters
    t, y, u = df['t'], df['y'], df['u']
    sys = tf([Tz,1],[Tw**2,2*Zeta*Tw,1])
    t, y_e = control.forced_response(sys,t,u)
    return y_e-y
```

Para ilustrar, fazendo um chute inicial dos dois parâmetros a ser estimados, avalia-se o resíduo no tempo.

```python
parameter_names = ['Zeta', 'Tw']
parameter_guess = [1, 0.025]

r = residuals(parameter_guess, df)
r.plot()
```

## Passo 3.1. Avaliando o "mérito" dos parâmetros estimados

Existem diversas formas de se avaliar o mérito, uma das mais usuais é o erro quadrático, sendo ele:

\begin{align*}
SOS & = \sum_k r_k^2
\end{align*}

O objetivo será portanto, encontrar os parâmetros que minimizem a soma dos erros quadráticos.

Definindo essas funções, observe que `sos` calcula a soma de todos os resíduos e eleva ao quadrado, já `best_fit` chama o algoritmo de minimização de Nelder-Mead implementado na biblioteca: `scipy.optimize.fmin`.

Referência:
Nelder, J.A. and Mead, R. (1965), “A simplex method for function minimization”, The Computer Journal, 7, pp. 308-313

```python
def sos(parameters, df):
    return sum(r**2 for r in residuals(parameters, df))

def best_fit(fcn, df, disp=1):
    return optimize.fmin(fcn, parameter_guess, args=(df,), disp=disp)

parameter_fit = best_fit(sos, df)

for name,value in zip(parameter_names, parameter_fit):
    print(name, " = ", round(value,4))
```

## Passo 4. Avaliar a qualidade do ajuste

```python
t, y, u = df['t'], df['y'], df['u']
Zeta,Tw = parameter_fit
sys = tf([Tz,1],[Tw**2,2*Zeta*Tw,1])
t1, y_e = control.forced_response(sys,t,u)
r = y_e-y
r.plot()
```

```python
# plotando a entrada, a saída medida e estimada no tempo
dff = pd.concat([df, pd.DataFrame(y_e,columns=['y_e'])],axis=1)
dff = dff.set_index('t')
dff.plot()
```
